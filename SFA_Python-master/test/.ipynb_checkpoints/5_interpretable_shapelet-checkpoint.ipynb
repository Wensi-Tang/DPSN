```python
import torch
import sys
import os
import sklearn
from sklearn import preprocessing
sys.path.append(os.getcwd()[:-5])
from src.timeseries.TimeSeriesLoader import uv_load
from  src.classification.BOSSEnsembleClassifier import *
import pickle
from scipy.sparse import csr_matrix
import numpy as np
from pyts.transformation import BOSS
from os.path import dirname
import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit
import pyts
from sklearn.feature_selection import f_regression
from sklearn.feature_selection import mutual_info_classif
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.metrics import mutual_info_score

class Shapelets():
    def __init__(self, window_size, step_size = 1,n_conf = 30, fft =False ):
        self.window_size = window_size
        self.step_size = step_size
        self.n_conf = n_conf
        self.fft = fft

    def forward(self,shapelets_featurex, X):
        X_candiate_window =self.pytorch_rolling_window(X_candiate)
        
        if self.fft:
            temp = torch.rfft(X_candiate_window,1,onesided=False)
            temp[0,0,int(self.n_conf/2):]=0
            X_candiate_window = torch.irfft(temp,1,onesided=False)
        
        X_candiate_window = torch.transpose(X_candiate_window, 2, 1)
        shapelets_distance = self.sliding_euclidean_distance(X_candiate_window, X)
        return X_candiate_window, shapelets_distance

    def pytorch_rolling_window(self,x):
        # unfold dimension to make our rolling window
        return x.unfold(1, self.window_size, self.step_size)

    def sliding_euclidean_distance(self,sample,samples ):
        sliding_sub_result = self.sliding_substraction(sample, samples)
        sliding_sub_result = torch.sum(sliding_sub_result**2, dim=1, keepdim=True)
        shapelets_distance,_ = torch.min(sliding_sub_result, dim= 3)
        return shapelets_distance

    def sliding_substraction(self, sample, samples):
        sample.unsqueeze_(-1)
        samples = self.pytorch_rolling_window(samples)
        samples = torch.transpose(samples, 2, 1)
        samples.unsqueeze_(-2)
        return samples-sample


def pick_and_remove_neightbor(index_list,shapelets_length,remove_ration= 1):
    feature_list=[]
    if remove_ration ==0:
        feature_list = [i for i in index_list]
    else:
        while len(index_list) != 0:
            ind = index_list[0]
            revelant_list = np.ceil(
                np.arange(ind - (shapelets_length * remove_ration), ind + (shapelets_length * remove_ration), 1))
            index_list = list(filter(lambda x: x not in revelant_list, index_list))
            feature_list.append(ind)
    return feature_list



    
def TSC_data_loader(dataset_name):
    Train_dataset = np.loadtxt(dirname(os.getcwd())+ '/datasets/UCRArchive_2018/'+ dataset_name+'/' + dataset_name + '_TRAIN.tsv')
    Test_dataset = np.loadtxt(dirname(os.getcwd())+ '/datasets/UCRArchive_2018/'+ dataset_name+'/'+ dataset_name + '_TEST.tsv')
    Train_dataset = Train_dataset.astype(np.float32)
    Test_dataset = Test_dataset.astype(np.float32)
    

    X_train = Train_dataset[:, 1:]
    y_train = Train_dataset[:, 0:1]

    X_test = Test_dataset[:, 1:]
    y_test = Test_dataset[:, 0:1]
    le = preprocessing.LabelEncoder()
    le.fit(np.squeeze(y_train, axis=1))
    y_train = le.transform(np.squeeze(y_train, axis=1))
    y_test = le.transform(np.squeeze(y_test, axis=1))
    
    #X_train = np.nan_to_num(X_train) 
    #X_test  = np.nan_to_num(X_test)

    return X_train, y_train, X_test, y_test
    
    
    

def BOSS_load_Superparameter(dataset_name):
    father_path = dirname(os.getcwd())+ '/test/BOSS_model/'+ dataset_name
    save_path = father_path + '/' + dataset_name + '_model.pkl'
    with open(save_path, 'rb') as input:
        models = pickle.load(input)
    return models

def get_super_parameter(dataset_name):
    
    models = BOSS_load_Superparameter(dataset_name)
    best_model = models[0]
    best_score = models[0][1]
    for i in range(len(models)):
        if models[i][1]>best_score:
            best_model = models[i]
    n_conf = best_model[2]
    windows = best_model[4]
    norm_mean = best_model[3].normMean   
    
    return n_conf, windows, norm_mean

def np_dft_filter(X,n_conf):
    temp = np.fft.fft(X)
    temp[int(n_conf/2):]=0
    result=np.fft.ifft(temp)
    return result
    
def load_shapelet_list(name):
    father_path = dirname(dirname(os.getcwd()))+'/time_series_proto/exp/test_accuracy_log_all_as_proto/' + name
    path = father_path + '/' + name + 'NN_list_log.txt'
    block=np.loadtxt(path)
    result = block
    return result

def save_to_log(sentence,dataset_name):
    father_path = './image_result/temp/' + dataset_name
    if not os.path.exists(father_path):
        os.makedirs(father_path)
    path = father_path+ '/' +dataset_name+ '.txt'
    with open(path, "a") as myfile:
        myfile.write(sentence+'\n')

def distance_reform(distance):
    temp =np.amin(distance)
    temp = np.expand_dims(temp,0)
    upp =(distance-temp)
    down =np.expand_dims((np.amax(distance)-np.amin(distance)),0)
    distance = upp/down
    return  distance 


name_list=[
    'ECG200',
  ]


    
#figbig = plt.figure()
sub_ind = 1
for name in name_list:
    print(name)
    fig= plt.figure()
    sub_ind = sub_ind + 1
    n_conf, windows, norm_mean = get_super_parameter(name)
    #print(n_conf, windows, norm_mean)
    windows = windows
    shapelets_length = windows
    X_train, y_train, X_test, y_test=TSC_data_loader(name)
    shapelets= Shapelets(shapelets_length,n_conf = n_conf,fft =True)
    n_feature =1
    f_score_pool = []
    shapelts_pool = []
    result_list = load_shapelet_list(name)
    class_number = np.max(y_train)+1

    sentence =str(class_number)
    save_to_log(sentence,name)
    print(result_list)
    result_list = result_list[0]
    for ind, i in enumerate(result_list):
        X_candiate = X_train[int(i):int(i) + 1, :]
        X_candiate = torch.tensor(X_candiate)
        batch = torch.tensor(X_train)
        X_candiate_window, shapelets_distance = shapelets.forward(X_candiate, batch)
        shapelts_pool.append(X_candiate_window)
        shapelets_distance.squeeze_(1)
        shapelets_distance = shapelets_distance.numpy()
        y_temp = np.ones(y_train.shape)*ind
        y_temp[np.argwhere(y_train != ind)] =1-ind
        f_test, P = f_regression(shapelets_distance, y_temp, center=True)
        f_test = np.asarray(f_test)
        ind_loc = np.argmax(f_test)
        print(class_number,ind)
        ax =fig.add_subplot(class_number, 2, ind+1)
        ax.axis([0, X_train.shape[1],np.amin(X_train) , np.amax(X_train)])
        ax.plot(X_train[int(i)])
        ax.set_title((str(int(i)+1)+'th train data'+', Lable='+str(ind)))
        ax.plot(np.linspace(ind_loc-1, ind_loc-1 + windows, windows), X_candiate_window[0,:,ind_loc,0].numpy(),label = 'DFT shapelet',color ='red',alpha=0.7)
    fig.subplots_adjust(top=(np.max(y_train)+1)*0.5, bottom=0.1,right=1, hspace=0.3, wspace= 0.35)
    plt.show()
    path = './image_result/'+name+'_1.pdf'
    fig.savefig(path,bbox_inches = 'tight')
    fig= plt.figure()
    ind_dic = {}
    for c in range(class_number):
        ind_dic[c] = np.argwhere(y_train == c)
        f_score_pool = []
        shapelets = Shapelets(shapelets_length, n_conf=n_conf,fft = False)
        for i in ind_dic[c]:
            X_candiate = X_train[int(i):int(i) + 1, :]
            X_candiate = torch.tensor(X_candiate)
            batch = torch.tensor(X_train)
            X_candiate_window, shapelets_distance = shapelets.forward(X_candiate, batch)
            shapelts_pool.append(X_candiate_window)
            # feature selection
            shapelets_distance.squeeze_(1)
            shapelets_distance = shapelets_distance.numpy()
            y_temp = np.ones(y_train.shape)
            y_temp[np.argwhere(y_train != ind)] = 0
            f_test, P = f_regression(shapelets_distance, y_temp, center=True)
            f_score_pool.append(f_test)

        f_score_pool = np.array(f_score_pool)
        index = np.unravel_index(np.argsort(f_score_pool.ravel())[-n_feature:], f_score_pool.shape)
        ind_train = ind_dic[c][index[0][0]]
        ind_loc = index[1][0]
        ax2 = fig.add_subplot(class_number, 2, c+1)
        ax2.axis([0, X_train.shape[1],np.amin(X_train) , np.amax(X_train)])
        ax2.set_title((str(int(ind_train[0])+1)+'th train data'+', Lable='+str(c)))
        ax2.plot(X_train[ind_train[0]])
        ax2.plot(np.linspace(ind_loc-1, ind_loc-1 + windows, windows), X_train[ind_train[0]][ind_loc:ind_loc + windows],label = 'shapelet',alpha=0.7)      
    fig.subplots_adjust(top=(np.max(y_train)+1)*0.5, bottom=0.1,right=1, hspace=0.3, wspace= 0.35)
    path = './image_result/'+name+'_2.pdf'
    fig.savefig(path,bbox_inches = 'tight')
    plt.show()





```
